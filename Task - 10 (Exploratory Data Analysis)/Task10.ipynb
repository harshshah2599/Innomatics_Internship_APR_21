### Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given to you.


```python
from scipy.stats import chi2
from scipy.stats import chi2_contingency
```

```python
sample=pd.crosstab(df['Specialization'],df['Gender'])
```

```python
chi2_contingency(sample)
chi2_test = chi2_contingency(sample)[0]
p_val = chi2_contingency(sample)[1]
df1 = chi2_contingency(sample)[2]
```

```python
# calculate chi critical

confidence = 0.90
alpha = 1 - confidence
chi2_critical = chi2.ppf(1-alpha,df1)
chi2_critical
```

```python
# Plot the chi2 distribution
x_min = 0
x_max = 100

# plot the graph with limits
x = np.linspace(x_min,x_max,100)
y = chi2.pdf(x,df1)
plt.xlim(x_min,x_max)
plt.plot(x,y)

# assign critical value
chi_critical_right = chi2_critical

# Shade the rejection part
x1 = np.linspace(chi_critical_right,x_max,100)
y1 = chi2.pdf(x1,df1)
plt.fill_between(x1,y1,color='red')
```

```python
# Conclusion with chi2 test
if(chi2_test>chi2_critical):
    print("Reject null hypothesis")
else:
    print("Failed to reject null hypothesis")
```

```python
# Conclusion with p-test
if (p_val<alpha):
    print("Reject null hypothesis")
else:
    print("Failed to reject null hypothesis")
```

```python
df1 = df[['Specialization','Gender']]
```

```python
sns.catplot( x='Specialization',
             kind="count", 
             hue="Gender", 
             height=10, 
             aspect=1.5, 
             data=df1)
plt.ylim(0,500)
plt.xticks(rotation=90)
```

### Is there a relationship between gender and specialisation? (i.e. Does the preference of Specialisation depend on the Gender?)
- Yes, completely. As we can observe in the above catplot the specializations such as computer engineering and its variants have more men as compared to women, women are about 50% less in number. 
- For specialization such Mechanical Engineering, it is surprisingly observed that the field is strongly dominated by men, and that women tend to choose against getting enrolled in such field as infered from the numbers that women are not even 10% of the men enrolled.
- In the field of computer science, the male and female numbers are approximately equal, followed by instrumentation and control engineering where the difference between the genders in very less, implying that both the specializations are equally preferred by both the genders. 
- For specialization such as biotechnology, the number of female students exceedes the male population, suggesting that biotechnology is a preference of females and comparitively less male students.
