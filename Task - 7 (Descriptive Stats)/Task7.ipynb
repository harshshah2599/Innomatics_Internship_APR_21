## 1. <u> Mean <u> :- 

<h3> Mean in statistics is nothing but the average of all the given numbers. It is calculated by summing up all the numbers and then dividing the sum by count of total numbers given. <h3>  

Formula:

<centre> <h2> $ Mean = \frac{1}{n} \sum_{i=i}^{n} x_{i} $ $ \ $ OR $  \  $ $ \frac{Sum \ of \ all \ the \ elements}{Total \ no \ of \ elements} $ <h2> <centre>

For Example:<br> 
Consider a given array => array_1=[5,7,1,6,2,4] So to compute the mean using the above formula we can write, 

Mean = $\frac{5+7+1+6+2+4}{6}$ = $\frac{25}{6}$

Now Let us calculate mean of 'Mthly_HH_Income' column from the imported dataset,

```python
df.mean()
```

```python
mean = df["Mthly_HH_Income"].mean()
print(mean)
```

## 2. <u> Median <u> :- 

<h3> Median in statistics is the middle element or value in a given sorted list of numbers.<br><br><ul style="list-style-type:circle">1. If the list contains odd number of elements, then the median is simply the middle value.<br>2. But if the list contains even number of elements, then the median is the average of the two middle values. </ul><h3>  

Formula:

<h2> $ Median = \frac{n+1}{2}^{th} \ (if \ n \ is \ odd) $ <br> $ Median = \frac{\frac{n}{2}^{th}+\frac{n+1}{2}^{th}}{2} \ (if \ n \ is \ even) $ <h2>

where, n is the number of elements in the list

For Example:<br> 
Consider a given array => array_1=[1,2,4,5,6,7] ; array_2=[5,7,9,12,15] So to compute the median using the above formula we can write, 

Median of array_1 = $\frac{4+5}{2}$ = $\frac{9}{2}$ <br> Median of array_2 = 9

Now Let us calculate median of 'Mthly_HH_Expense' column from the imported dataset,

```python
df.median()
```

```python
median = df["Mthly_HH_Expense"].median()
print(median)
```

## 4. <u> Variance <u> :- 

<h3> The term variance refers to a statistical measurement of the spread between numbers in a data set. More specifically, variance measures how far each number in the set is from the mean and thus from every other number in the set. It is often depicted by this symbol: σ2 <h3>

Formula:

<h2> $ \sigma^2 = \frac{\displaystyle\sum_{i=1}^{n}(x_i - \mu)^2} {n} $ <h2>
 

where, 
x represents each value in the dataset.
$ \mu $ stands for mean.
n is the total no of values.

For Example:<br> 
Consider a given array => array_1=[5,7,1,6,2,4] So to compute the variance using the above formula we can write, 

Variance = $ \sigma^2 = \frac{\displaystyle\sum_{i=1}^{6}(x_i - 4.16)^2} {6} $

Now Let us calculate variance of 'Emi_or_Rent_Amt' column from the imported dataset,

```python
variance = df["Emi_or_Rent_Amt"].var()
print(variance)
```

## 5. <u> Standard Deviation <u> :- 

<h3> A standard deviation is a statistic that measures the dispersion of a dataset relative to its mean. The standard deviation is calculated as the square root of variance by determining each data point's deviation relative to the mean. <h3>

Formula:

<h2> $ \sigma = \sqrt\frac{\displaystyle\sum_{i=1}^{n}(x_i - \mu)^2} {n} $ <h2>


where, 
x represents each value in the dataset.
$ \mu $ stands for mean.
n is the total no of values.

For Example:<br> 
Consider a given array => array_1=[5,7,1,6,2,4] So to compute the standard deviation using the above formula we can write, 
<h2> $ \sigma = \sqrt\frac{\displaystyle\sum_{i=1}^{6}(x_i - \mu)^2} {6} $ <h2>


Now Let us calculate standard deviation of 'Emi_or_Rent_Amt' column from the imported dataset,

```python
sd = df["Emi_or_Rent_Amt"].std()
print(sd)
```

## 7. <u> Normal Distribution <u> :- 

<h2> A normal distribution, sometimes called the bell curve, is a distribution that occurs naturally in many situations. A Normal Distribution is also known as a Gaussian distribution or famously Bell Curve. People use both words interchangeably, but it means the same thing. It is a continuous probability distribution. <h2>

```python
import numpy as np
import matplotlib.pyplot as plt
 
# Creating a series of data of in range of 1-50.
x = np.linspace(1,50,200)
 
#Creating a Function.
def normal_dist(x , mean , sd):
    prob_density = (np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density
 
#Calculate mean and Standard deviation.
mean = np.mean(x)
sd = np.std(x)
 
#Apply function to the data.
pdf = normal_dist(x,mean,sd)
 
#Plotting the Results
plt.plot(x,pdf , color = 'red')
plt.xlabel('Data points')
plt.ylabel('Probability Density')
```

## 10. <u> Effect on Mean, Median and Mode due to Skewness <u> :- 

<h3>In a summary, generally if the distribution of data is skewed to the left, the mean is less than the median, which is often less than the mode. If the distribution of data is skewed to the right, the mode is often less than the median, which is less than the mean.<h3>



For Example:

Consider the data - [4;5;6;6;6;7;7;7;7;8]


The mean is 6.3, the median is 6.5, and the mode is seven. Notice that the mean is less than the median, and they are both less than the mode. The mean and the median both reflect the skewing, but the mean reflects it more so.

Consider another data - [6;7;7;7;7;8;8;8;9;10]


The mean is 7.7, the median is 7.5, and the mode is seven. Of the three statistics, the mean is the largest, while the mode is the smallest. Again, the mean reflects the skewing the most.

## 11. <u> Explain QQ Plot and show the implementation of the same <u> :- 

 <h3>A Q–Q plot is a plot of the quantiles of two distributions against each other, or a plot based on estimates of the quantiles. The pattern of points in the plot is used to compare the two distributions.
    Q-Q plots are used to find the type of distribution for a random variable whether it be a Gaussian Distribution, Uniform Distribution, Exponential Distribution or even Pareto Distribution, etc. In general, we are talking about Normal distributions only because we have a very beautiful concept of 68–95–99.7 rule which perfectly fits into the normal distribution<h3>
        

```python
import pylab 
import scipy.stats as stats
for var in ['Mthly_HH_Income', 'Mthly_HH_Expense', 'No_of_Fly_Members', 'Emi_or_Rent_Amt', 'Annual_HH_Income', 'No_of_Earning_Members']:
    
    plt.figure(figsize=(15,6))
    plt.subplot(1, 2, 2)
    stats.probplot(df[var], dist="norm", plot=pylab)
    plt.title('Q-Q plot of '+ var)

    plt.show()
```
